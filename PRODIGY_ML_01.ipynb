```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

```

```python

# Load the dataset
Dst = pd.read_csv("kc_house_data.csv")

# Drop unnecessary columns 'date' and 'id'
Dst = Dst.drop(['date', 'id'], axis=1)

# Select the feature 'bedrooms'
X = Dst[['bedrooms']]

# Select the target variable 'price'
Y = Dst['price']

# Split the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)

# Create a Linear Regression model
Linear_Regression = LinearRegression()

# Train the model
Linear_Regression.fit(X_train, Y_train)

# Predict the target variable for the test set
Y_predict = Linear_Regression.predict(X_test)

print(Dst.head())
```

```python
# Print the predicted values
print("Predicted")
print(Y_predict)

# Print the coefficient and intercept of the model
print("Coefficient : ")
print(Linear_Regression.coef_)
print("Intercept : ")
print(Linear_Regression.intercept_)
```

```python
# Data visualization
plt.figure(figsize=(10, 6))
plt.scatter(X_test, Y_test, alpha=0.5, edgecolors='k', label='Actual Prices')
plt.plot(X_test, Y_predict, color='red', linewidth=2, label='Predicted Prices')
plt.xlabel("Number of Bedrooms")
plt.ylabel("Prices")
plt.title("Number of Bedrooms vs Prices")
plt.legend()
plt.show()
```
